```python
# Task Instructions
# 1. The application should ask the user whether they want to continue or exit.
# 2. If the user selects Continue, the menu options should be displayed again.
# 3. This process should repeat until the user selects 0 (Exit).
# 4. Implement logic to limit the number of transactions to 3 per day.
# 5. Once the limit is reached, the application should not allow further transactions.
# 6. After each transaction, display the number of currency notes (How many 500, 200 and 100).
class Bank:
    acbal = 10000
    txn_count = 0  

    def showNotes(self, amt):
        n500 = amt // 500
        amt = amt % 500

        n200 = amt // 200
        amt = amt % 200

        n100 = amt // 100

        print("Notes:")
        print("500 :", n500)
        print("200 :", n200)
        print("100 :", n100)

    def withdraw(self):
        if self.txn_count >= 3:
            print("Transaction limit reached for today!")
            return

        amt = int(input("Enter your withdraw amount :"))

        if amt % 100 == 0:
            if amt <= 20000:
                if amt <= self.acbal:
                    self.acbal = self.acbal - amt
                    self.txn_count += 1
                    print("Available bal : ", self.acbal)
                    self.showNotes(amt)
                else:
                    print("insuffient fund")
            else:
                print("withdraw limit is 20k only")
        else:
            print("Please enter multiples of 100 only")

    def deposite(self):
        if self.txn_count >= 3:
            print("Transaction limit reached for today!")
            return

        amt = int(input("Enter your deposite amount: "))

        if amt % 100 == 0:
            self.acbal = self.acbal + amt
            self.txn_count += 1
            print("Avalaible bal :", self.acbal)
            self.showNotes(amt)
        else:
            print("Please enter multiples of 100 only")

    def viewOptions(self):
        while True:
            print("\n1. deposite")
            print("2. withdraw")
            print("3. bal Enquiry")
            print("0. EXIT")

            option = int(input("Choose your option: "))

            if option == 1:
                self.deposite()
            elif option == 2:
                self.withdraw()
            elif option == 3:
                print("Bal Enquiry :", self.acbal)
            elif option == 0:
                print("Thank you, visit again")
                break
            else:
                print("Invalid option")

            if self.txn_count >= 3:
                print("\nYou have reached the maximum of 3 transactions for today.")
                break

            ch = input("\nDo you want to continue? (y/n): ").lower()
            if ch != 'y':
                print("Thank you, visit again")
                break


obj = Bank()
obj.viewOptions()
```
